```python
#I could not automate scraping for this outlet so I manually webscraped 5 pages of search results for "campus protest". 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the New York Post search results
url = "https://nypost.com/search/campus+protest/page/13/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the response with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all headlines in the search results
    headlines = soup.find_all('h3', class_='headline')
    
    # Create a list to store the headlines
    headlines_list = []

    # Loop through the headlines and add them to the list
    for i, headline in enumerate(headlines):
        if i >= 100:  # Only get the first 100 headlines
            break
        headlines_list.append(headline.get_text(strip=True))

    # Create a DataFrame from the list of headlines
    df = pd.DataFrame(headlines_list, columns=['Headline'])
    
    # Display the DataFrame
    print(df)
```

```python
#Now I compiled all the headlines I scraped into one dataframe. 
dataframes = [df, df_2, df_3, df_4, df_5]

all_df = pd.concat(dataframes, ignore_index=True)

all_df = all_df.drop(index=0)
all_df = all_df.reset_index(drop=True)
all_df
```

```python
#I ran the sentiment analysis
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from io import StringIO  # Add this import

# Download necessary NLTK data
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis on a text
def analyze_sentiment(all_df):
    return sia.polarity_scores(all_df)

# Apply sentiment analysis to the 'headline' column
all_df['sentiment'] = all_df['Headline'].apply(analyze_sentiment)

# Extract compound sentiment score
all_df['compound_sentiment'] = all_df['sentiment'].apply(lambda x: x['compound'])

# Calculate average sentiment
average_sentiment = all_df['compound_sentiment'].mean()

print("Sentiment analysis results:")
print(f"Average sentiment score: {average_sentiment:.4f}")

# Display individual headline sentiments
print("\nIndividual headline sentiments:")
for index, row in all_df.iterrows():
    print(f"Headline: {row['Headline']}")
    print(f"Sentiment: {row['compound_sentiment']:.4f}")
    print()
```
